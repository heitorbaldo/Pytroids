```python
import itertools
import numpy as np
import networkx as nx

from pytroids.graphic_matroids import *
from pytroids.dual_matroids import *
```

## Mininal Examples

```python
#Example: |V| = 4
A = np.array([
    [0, 1, 1, 0],
    [1, 0, 1, 0],
    [1, 1, 0, 1],
    [0, 0, 1, 0]
])

G = nx.from_numpy_array(A)
nx.draw_networkx(G, with_labels=True)
```

```python
M = GraphicMatroid(A)

print("Edges:", M.edges)
print("Rank:", M.rank(M.edges))
print("Bases:", M.bases())
print("Circuits:", M.circuits())
```

```python
#Dual matroid

M = GraphicMatroid(A)
M_star = DualMatroid(M)

print("Primal bases:", M.bases())
print("Dual bases:", M_star.bases())
print("Dual rank:", M_star.rank())
```

```python
#Example: |V| = 4 (NetworkX version)

G = nx.from_numpy_array(A)
M = GraphicMatroidNX(G)

print("Edges:", M.edges)
print("Rank:", M.rank())
print("Bases:", M.bases())
print("Circuits:", M.circuits())
```

```python
#Dual matroid

M = GraphicMatroidNX(G)
M_star = DualMatroid(M)

print("Primal bases:", M.bases())
print("Dual bases:", M_star.bases())
print("Dual rank:", M_star.rank())
```

## Examples with Random Graphs

```python
n = 10
p = 0.2

G_er = nx.gnp_random_graph(n, p, seed=None, directed=False)

print(G_er.number_of_edges())
nx.draw_circular(G_er, with_labels=True)
```

```python
M = GraphicMatroidNX(G_er)

#print("Edges:", M.edges)
print("Rank:", M.rank())
print("Bases:", M.bases())
print("Circuits:", M.circuits())
```
